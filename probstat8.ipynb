```python
import pandas as pd
#from pandas.tools import plotting 
import matplotlib.pyplot as plt 
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier,export_graphviz 
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
data_rombon = pd.read_csv('prob/houseprices.csv') 
data_rombon.head(5)
```

```python
col_list=['Price']
numhouse = data_rombon[data_rombon.columns[data_rombon.columns.isin(col_list)]]
plt.figure(figsize=(10,5))
numhouse.boxplot(sym='r*', grid=False) 
plt.show()
```

```python
plt.figure(figsize=(15,5)) 
plt.subplot(121)
data_rombon['Price'].plot.hist(bins=10, title='Price') 
plt.show()
```

```python
col_list=['Price', 'SqFt']
numhouse = data_rombon[data_rombon.columns[data_rombon.columns.isin(col_list)]] 
numhouse.plot.scatter(x='SqFt', y='Price')
```

```python
col_list=['Price', 'Bedrooms']
numhouse = data_rombon[data_rombon.columns[data_rombon.columns.isin(col_list)]] 
plt.figure(figsize=(15,5))
numhouse.boxplot(by='Bedrooms')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_rombon['Bedrooms'], data_rombon['Price'], alpha=0.5)
plt.title('Scatter Plot Harga Rumah Berdasarkan Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data_rombon.boxplot(column='Price', by='Bathrooms')
plt.title('Grouped Boxplot Berdasarkan Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
